## Permutations
1. How many ways can you arrange the numbers 1, 2, 3, 4 and 5? 
**5! = 5*4*3*2*1 = 240**
2. How many ways can you arrange 1, 1, 2, 3, 4? 
**5!/2! = (5*4*3*2)/2 = 60**
3. How many ways can you arrange two 3s and three 5s? 
**5!/3!2! = 10**


## Combinations
1. How many different poker hands (5 cards) can you have? A deck holds 52 cards. 
** 52!/5!*(52-5)! = 52!/ 5! * 47! = (52*51*50*49*48)/(5*4*3*2*1) = 2,598,960 **

## Probability of an event
1. In a deck of cards (52 cards), what’s the probability of picking a queen? **4/52 = 1/13 = 7.7%**
A heart? **13/52 = 1/4 = 25%**
Of picking a card that’s not a queen nor a heart? **1-36/52 = 1 - 0.692 = 30.8% **
2. If I do not replace the cards, what is the probability of picking 2 kings? ** 4/52*3/51 = 12/2652 = 12/2652 = .0045**
4 diamonds? ** 13/52*12/51*11/50*10/49 = 1716/649740 = 286/108290 **
How do these probabilities evolve if I replace the cards after each draw?
** 2 kings: 4/52*4/52 = 4/676, 4 diamonds: 13/52*13/52*13/52*13/52 = 4/52**

**In general, the number of ways to select k distinct items without replacement from n items in any order is (nk) = n!/(n-k)!k!. You can show that here are k! times more permutations than combinations when you choose k items from n items without replacement. That's because the formula for a permutation is n!/(n-k)!, while the formula for a combination is n!/(n-k)!k!. If you take k! times the combination formula, you get the permuation formula. That means it takes k! combinations to equal one permuation (which is true because with a combination, you don't care hoe those k items are arranged, and with a permutation, you do)**

### Conditional probability
1. What is the probability that the total of two dice is less than four, knowing that the first die is a 2?
**(1,2), (2,2), (2,1) = 3/36 = 1/12**
2. 90% of candidates to a Web developer position can code both in Javascript and HTML. 70% of these candidates can code in Javascript and 50% can code in HTML. What is the probability that a candidate can code in HTML knowing that he can code in Javascript?
**90% - 70% = 20%; 90% - 50% = 40%; 90% - 20% - 40% = 30%**

2. Extra Credit: A jar contains red and blue marbles. You draw two marbles one after the other without replacing the first marble in the jar. You know that:
  • The probability of selecting a blue marble and then a red marble is 30%.
  • The probability of selecting a red marble on the first draw is 50%.
You first draw a red marble. What is the probability of selecting a blue marble on the second draw? 

     **P(blue, red)=30%
     P(red)= 50%
     P(blue) = 50%
     P(red|blue) = P(blue|red) * P(red) / P(blue)
     P(red|blue) = .3 * .5 /.5
     P(red|blue) = 30%**
